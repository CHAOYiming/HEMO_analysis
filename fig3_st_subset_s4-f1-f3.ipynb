```python
import scanpy as sc
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, facecolor='white')
sc.settings.verbosity = 3
```

```python
# load slice4 adata
adata = sc.read('/home/yiming/vio_analysis/ST/slice4/sc_slice4_1217.h5ad')
# load RCTD
cellpred = pd.read_csv('/usersdata/yiming/VIO/Ranalysis/RCTD_june/slice4_norm_wei_top2.csv', index_col=0)
cellpred
adata.obs = adata.obs.drop(columns=['id'])
adata.obs = pd.DataFrame.join(adata.obs, cellpred['top1'])
# subset frag1
frag1 = pd.read_csv('frag1.csv', index_col=0)
adata_subset1 = adata[frag1.index].copy()
```

```python
adata_subset1.uns['top1_colors'] = ['#477AB4', '#FA9F9A', '#9B4D9E', '#62ACE0', '#BF9BC9', '#911740']
```

```python
sc.pl.spatial(adata_subset1, color=['top1'], size=1.5, title=['RCTD'])
```
