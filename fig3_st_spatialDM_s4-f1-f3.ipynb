```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import colors
import matplotlib.pyplot as plt 
```

```python
matplotlib .rcParams['figure.dpi'] = 320
plt.style.use('seaborn-white')
```

```python
# load LR pairs
LR_pairs = pd.read_csv('/home/yiming/vio_analysis/ST/spatialDM/LR_frag1_pairs.csv')
```

```python
# load spatial coordinate
coord = pd.read_csv('/home/yiming/vio_analysis/ST/spatialDM/D1_frag3_tissue_positions_list.csv', 
                    index_col=0, header = None)
coord = coord.loc[coord[1] == 1, [4, 5]]
coord.columns = ['x', 'y']
```

```python
# subset EPO
EPO_pairs = LR_pairs.loc[:, LR_pairs.columns.str.startswith('EPO')]
```

```python
# frag1 EPO
plt.style.use('seaborn-white')
plt.figure(figsize=(2.2, 1.5))
for i,c in enumerate(EPO_pairs.columns):
    plt.scatter(coord.y, coord.x*(-1), 
                marker = 'o', c=EPO_pairs[c], cmap='Reds', alpha=0.7, linewidth=0)
    plt.axis('off')
    plt.title(c, fontsize=7)
```

```python
# load LR pairs
LR_pairs = pd.read_csv('/home/yiming/vio_analysis/ST/spatialDM/LR_frag3_pairs.csv')
```

```python
# subset signaling
WNT_pairs = LR_pairs.loc[:, LR_pairs.columns.str.startswith('WNT')]
BMP_pairs = LR_pairs.loc[:, LR_pairs.columns.str.startswith('BMP')]
# calculate mean 
WNT_mean = WNT_pairs.mean(axis=1)
WNT_mean = pd.DataFrame(WNT_mean, columns=['WNT_mean'])
BMP_mean = BMP_pairs.mean(axis=1)
BMP_mean = pd.DataFrame(BMP_mean, columns=['BMP_mean'])
```

```python
# frag3 WNT_mean
plt.style.use('seaborn-white')
plt.figure(figsize=(1, 1.1))
for i,c in enumerate(WNT_mean.columns):
#     plt.subplot(5, 4, i+1)
    plt.scatter(coord.y, coord.x*(-1),  
                marker = 'o', c=WNT_mean[c], cmap='Reds', alpha=0.7, linewidth=0, s=6)
    plt.axis('off')
    plt.title(c, fontsize=7)
```

```python
# frag3 BMP_mean
plt.style.use('seaborn-white')
plt.figure(figsize=(1, 1.1))
for i,c in enumerate(BMP_mean.columns):
#    plt.subplot(6, 5, i+1)
    plt.scatter(coord.y, coord.x*(-1),  
                marker = 'o', c=BMP_mean[c], cmap='Reds', alpha=0.7, linewidth=0, s=6)
    plt.axis('off')
    plt.title(c, fontsize=7)
```
